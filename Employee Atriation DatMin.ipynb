```python
import pandas as pd 
import numpy as np
import xlrd as xl 
from pandas import ExcelWriter
from pandas import ExcelFile 
df=pd.read_excel("Employee Attrition.xlsx")
df.head()
```

```python
#1.a
df2 = df[df["Age"]>=21]
df2.head()
```

```python
#1.b tidak dilakukan apapun karena saya belum pernah bekerja
#1.c 
df3 = df2.dropna(axis='rows')
df3.isnull().sum()
```

```python
df4a =df3[df3["Gender"]== ("Female")]
df4b =df3[df3["Gender"]== ("Male")]
df4 = [df4a , df4b]
df4=pd.concat(df4)
#df4=df4[df4["MaritalStatus"]==("Maried","Single","Divorced")] df2 = df[df["Age"]>=21]
```

```python
df4
```

```python
#df4=df4[df4["MaritalStatus"]==("Maried","Single","Divorced")] df2 = df[df["Age"]>=21]
df5a = df4[df4["MaritalStatus"]==("Married")]
df5b = df4[df4["MaritalStatus"]==("Single")]
df5c = df4[df4["MaritalStatus"]==("Divorced")]
df5 = [df5a , df5b , df5c]
df5 = pd.concat(df5)
df5.count()
```

```python
df5["Attrition"]= df5["Attrition"].map({"Yes":1, "No" : 0})
```

```python
df5
```

```python
import pandas as pd 

dfY = df5["Attrition"]
dfX = df5.drop(columns=["Attrition"])
```

```python
X=pd.get_dummies(dfX)
Y=dfY
```

```python
X.head()
```

```python
from sklearn.cross_validation import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2)
```

```python
#======== no 3 =========

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 10)
model.fit (xtrain, ytrain)
```

```python
import graphviz
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
from IPython.display import display


dot_data = export_graphviz(model, out_file=None, feature_names=list(xtrain), class_names=['0','1'], filled=True, rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data)

display(Image(graph.create_png()))
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest,ytest)
```

```python
#mencari deph optimum
i=1
maks=0
deph = 0
while (i <= 30) :
    model = DecisionTreeClassifier(max_depth = i)
    model.fit (xtrain, ytrain)
    
    nilai = model.score(xtest,ytest)
    if (nilai >maks):
        maks= nilai
        deph= i
    i = i + 1

model = DecisionTreeClassifier(max_depth = deph)
model.fit (xtrain, ytrain)

dot_data = export_graphviz(model, out_file=None, feature_names=list(xtrain), class_names=['0','1'], filled=True, rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data)

display(Image(graph.create_png()))



        
```

```python
print (deph)
print (model.score(xtrain, ytrain))
print (model.score(xtest,ytest))
```

```python
#kesimpulan:
#deph optimal adalah 3 dengan :
#  Score dari train : 0.8638412984670875
#  Score dari test :0.8201438848920863
#  Dengan menggunakan decision tree yang terbentuk untuk menghindari pegawai yang akan meninggalkan pekerjaannya, kita dapat menyeleksi dengan menghindari org yang :
#   1. Orang yang tidak overtime, monthly income <= 1683.5 , dan job role bukan research scientist
#   2. Orang yang overtime, monthly income <= 3779 dan job statisfication <= 1,5
#  Selain 2 tipe orang tersebut maka akan cenderung untuk tidak keluar dari pekerjaan.
```
